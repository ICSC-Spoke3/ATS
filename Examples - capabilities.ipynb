# Anomaly detector capabilities

This example shows the capabilities of some models.

```python
from ats.anomaly_detectors.base import AnomalyDetector
from ats.anomaly_detectors.naive.minmax import MinMaxAnomalyDetector
from ats.anomaly_detectors.stat.periodic_average import PeriodicAverageAnomalyDetector
from ats.anomaly_detectors.stat.robust import NHARAnomalyDetector
from ats.anomaly_detectors.ml.linear_regression import LinearRegressionAnomalyDetector
from ats.anomaly_detectors.ml.ifsom import IFSOMAnomalyDetector
from ats.anomaly_detectors.dl.lstm import LSTMAnomalyDetector
```

```python
anomaly_detectors = [MinMaxAnomalyDetector,
                     PeriodicAverageAnomalyDetector,
                     NHARAnomalyDetector,
                     LinearRegressionAnomalyDetector,
                     IFSOMAnomalyDetector,
                     LSTMAnomalyDetector]
```

```python
for anomaly_detector in anomaly_detectors:
    print()
    print(anomaly_detector.__name__)
    print(anomaly_detector.capabilities)
```
