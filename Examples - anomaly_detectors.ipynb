```python
from ats import logger
logger.setup('INFO')
```

```python
from ats.utils import generate_timeseries_df, plot_timeseries_df
```

```python
timeseries_df = generate_timeseries_df(entries=100, variables=2, freq='H')
```

```python
from ats.anomaly_detectors import MinMaxAnomalyDetector
```

```python
anomaly_detector = MinMaxAnomalyDetector()
```

```python
timeseries_df_scored = anomaly_detector.apply(timeseries_df)
```

```python
plot_timeseries_df(timeseries_df_scored, image=True)
```
