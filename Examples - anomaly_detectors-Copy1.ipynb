```python
from ats import logger
import pandas as pd
logger.setup('INFO')
```

```python
from ats.utils import generate_timeseries_df, plot_timeseries_df
timeseries_df = generate_timeseries_df(entries=100, variables=2, freq='H')

from ats.anomaly_detectors.dl.lstm import LSTMAnomalyDetector
from ats.anomaly_detectors.stat.periodic_average import PeriodicAverageAnomalyDetector
from ats.anomaly_detectors.ml.linear_regression import LinearRegressionAnomalyDetector
from ats.anomaly_detectors.naive import MinMaxAnomalyDetector
#detector = PeriodicAverageAnomalyDetector()
detector = LSTMAnomalyDetector()
#detector = LinearRegressionAnomalyDetector()
#detector = MinMaxAnomalyDetector()
detector.fit(timeseries_df)
result_df = detector.apply(timeseries_df)
plot_timeseries_df(result_df, image = True)
```
